```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
```

### Loading DATA

```python
data = load_iris()
```

```python
X = pd.DataFrame(data['data'],columns=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width'])
y = pd.DataFrame(data['target'],columns=['Target'])
```

```python
X.columns
```

### Scatter plot i.e grouping the flowes based on sepal l & w

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
color = np.array(['red','green','blue'])
plt.scatter(X['Sepal_Length'],X['Sepal_Width'],c=color[y.Target])
plt.title('REAL')
#kmeans
plt.subplot(1,3,2)
km = KMeans(n_clusters=3)
km.fit(X)
pred = km.labels_
plt.title('K_Means')
plt.scatter(X['Sepal_Length'],X['Sepal_Width'],c=color[pred])
#GM
#scaling
std = StandardScaler()
xs = std.fit_transform(X)
plt.subplot(1,3,3)
plt.title('GaussiaMixture')
gm = GaussianMixture(n_components=3)
gm.fit(xs)
pred = gm.predict(xs)
plt.scatter(X['Sepal_Length'],X['Sepal_Width'],c=color[pred])
```

```python
#GaussianMixture model exactly matches with real target than kmeans
```
